```r
install.packages("ggplot2")
library(ggplot2)

data(mtcars)
mtcars_model <- lm(mpg ~ wt, data = mtcars)
summary(mtcars_model)

ggplot(mtcars, aes(x = wt, y = mpg)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", color = "red", se = FALSE) +
  ggtitle("MPG vs Weight")

ggsave("mtcars_regression.png", width = 6, height = 4)

data(iris)
iris_setosa <- subset(iris, Species == "setosa")
iris_model <- lm(Sepal.Length ~ Petal.Length, data = iris_setosa)
summary(iris_model)

ggplot(iris_setosa, aes(x = Petal.Length, y = Sepal.Length)) +
  geom_point(color = "purple") +
  geom_smooth(method = "lm", color = "orange", se = FALSE) +
  ggtitle("Sepal Length vs Petal Length (Setosa)")

ggsave("iris_regression.png", width = 6, height = 4)
```
